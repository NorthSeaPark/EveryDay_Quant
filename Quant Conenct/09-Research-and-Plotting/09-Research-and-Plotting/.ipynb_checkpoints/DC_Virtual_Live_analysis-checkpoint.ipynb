```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Marcial/Downloads/2020-2021_Bookings-by-Modality_WashingtonDC.xlsx - Data.csv")
```

```python
def convert(year,month):
    
    return month+year[-2:]
```

```python
df['Timestamp'] = df[['Year','Month']].apply(lambda df:convert(df['Year'],df['Month']),axis=1)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'],format='%b%y')
```

```python
df = df.set_index('Timestamp')
```

```python
df['Product Revenue'] = df['Product Revenue'].replace('[\$,]', '', regex=True).astype(float)
```

```python
gg = df.groupby(['Timestamp','Delivery Method']).sum()['Product Revenue']
```

```python
gg = pd.DataFrame(gg)
```

```python
gg = gg.reset_index()
```

```python
gg
```

```python
gg = pd.pivot(gg,index='Timestamp',columns='Delivery Method',values='Product Revenue')
```

```python
import matplotlib.pyplot as plt
```

```python
modalities = gg[['Live','VOD','Virtual']]
```

```python
fig,ax = plt.subplots(figsize=(10,3),dpi=200)
modalities['VOD'].plot(ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(10,3),dpi=200)
modalities[['Live','Virtual']].pct_change().plot(ax=ax)
```

```python
modalities[['Live','Virtual','VOD']].corr()
```

```python
modalities[['Live','Virtual','VOD']].pct_change().corr()
```

```python
modalities['Total'] = modalities['Live']+modalities['Virtual']+modalities['VOD']
```

```python
(modalities['Virtual']/modalities['Total']).plot()
(modalities['Live']/modalities['Total']).plot()
```
