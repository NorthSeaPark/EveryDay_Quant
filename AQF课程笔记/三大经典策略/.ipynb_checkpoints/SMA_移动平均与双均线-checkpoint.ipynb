## 移动平均与双均线模型

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
import matplotlib as mpl
import numpy as np
import pandas as pd
mpl.rcParams['font.family'] = 'serif'               #解决一些字体显示乱码问题
import warnings; warnings.simplefilter('ignore')       #忽略警告信息；
```

```python
df = pd.read_csv("../data/PTR.csv")
df.head(10)
```

```python
df.set_index('Date', inplace = True)
df.head()
```

```python
#计算前20天的平均值
df['SMA_20'] = df['Close'].rolling(20).mean()
#计算前60天的平均值
df['SMA_60'] = df['Close'].rolling(60).mean()
df.tail(7)
```

```python
df[['Close','SMA_20','SMA_60']].plot(figsize = (10,6))
```

```python
#计算股票return
#公式：log(今日收盘价/昨日收盘价)
#可使用向量计算代替循环
df['returns'] = np.log(df['Close'] / df['Close'].shift(1))
```

```python
#计算每日盈亏比
#今日收盘价/昨日收盘价 - 1
df['return_dis'] = df['Close'] / df['Close'].shift(1) - 1
df['return_dis2'] = df['Close'].pct_change()
```

```python
#策略精髓：如果二十日平均 > 六十日平均 则做多 否则 做空
df['position'] = np.where(df['SMA_20'] > df['SMA_60'], 1 , -1)
```

### SMA策略

```python
data = pd.read_csv("../data/PTR.csv")
data.set_index('Date', inplace = True)
data.rename(columns={'Close':'Price'}, inplace = True)
data.head(7)
```

```python
data['SMA_10'] = data['Price'].rolling(10).mean()
data['SMA_60'] = data['Price'].rolling(60).mean()
```

```python
data.tail()
```

```python
data[['Price', 'SMA_10', 'SMA_60']].plot(title='10 & 60 SMA', figsize = (10,6))
```

### 策略思路

```python
data['Position'] = np.where(data['SMA_10']>data['SMA_60'],1,-1)
```

```python
data.dropna(inplace = True)
```

```python
data['Position'].plot(ylim=[-1.1, 1.1], title = 'Market Positioning',figsize=(10,8))
```

### 计算策略年化收益

```python
data['Returns'] = np.log(data['Price']/data['Price'].shift(1))
data['Returns'].hist(bins=35)
```

```python
#计算回测收益
data['Strategy'] = data['Position'].shift(1) * data['Returns']
```

```python
data.dropna(inplace=True)
```

```python
data.head(20)
```

```python
data[['Returns']].sum()
```

```python
data[['Strategy']].sum()
```

```python
data[['Returns', 'Strategy']].cumsum().apply(np.exp).plot(figsize=(10, 6))  
```

### 策略优化的另一种思路

```python
data1 = data[['Price','SMA_10','SMA_60']]
```

```python
data1.tail()
```

```python
data1[['Price','SMA_10','SMA_60']].plot(grid=True, figsize = (10,8))
```

```python
#我们现在设定一个 十日线减去六十日线 的数值
data1['10-60'] = data1['SMA_10'] - data1['SMA_60']
data1['10-60'].tail()
```

```python
SD = 20 #阀值
#如果十日均线比六十日均线 大 20 则做多 否则平仓
#接着，如果十日均线比六十日均线 小 20 则做空 否则继续之前的仓位状态
data1['Position'] = np.where(data1['10-60']>SD, 1, 0)
data1['Position'] = np.where(data1['10-60']<SD, -1, data1['Position'])

#例子： data1['10-60']的数值
# 21 30 45 19 -19 -21 -22 -10
# 做多 持仓做多 持仓做多 平仓 平仓 持仓做空 持仓做空 平仓
```
